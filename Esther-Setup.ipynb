## Words with same consonants but different lexical meanings
The following query sercahes Estehr chpater six for words which have the same consonants but having different lexical meanings within a range of 20 words.

```python
DiffLexSameCons = '''
chapter book=Esther chapter=6
    w1:word
    w2:word
w1 < w2
w1 <20: w2
w1 .lex_cons=lex_cons. w2
w1 .lex#lex. w2
'''
DiffLexSameCons  = BHS.search(DiffLexSameCons)
BHS.table(DiffLexSameCons, start=1, end=10, condensed=False,  colorMap={2: 'magenta', 3:'cyan'})
```

The following is the same query but for the whole book of Esther.

You may view the full list of 14 results by increasing the "end" number.

```python
SameLex = '''
chapter book=Esther
    w1:word
    w2:word
w1 < w2
w1 <20: w2
w1 .lex_cons=lex_cons. w2
w1 .lex#lex. w2
'''
SameLex  = BHS.search(SameLex)
BHS.table(SameLex, start=1, end=5, condensed=False,  colorMap={2: 'magenta', 3:'cyan'})
```

## Parallelism with similar first word
The following query is more commonly run for poetry.

It compares the first four phrsaes within a clause for a similar function and in similar arrangement to the clasue that follows it.

In addition, it includes an option requiring the first word of the the first phrase of both clauses to share the same consonants.

```python
Parallelism1 = '''
chapter book=Esther
    c1:clause
        =: p1:phrase
            w1:word
        <: p2:phrase
        <: p3:phrase
        <: p4:phrase
    c2:clause
        =: pA:phrase
            wA:word
        <: pB:phrase
        <: pC:phrase
        <: pD:phrase

c1 <: c2
p1 .function=function. pA
p2 .function=function. pB
p3 .function=function. pC
p4 .function=function. pD
w1 .g_cons=g_cons. wA
'''
Parallelism1  = BHS.search(Parallelism1)
BHS.table(Parallelism1, start=1, end=10, condensed=False, colorMap={1: 'black', 2: 'orange', 3: 'magenta', 4: 'yellow', 5: 'green', 6: 'yellow', 7: 'green', 8: 'orange',9: 'magenta',10: 'grey'})
```

This is a similar query but without the requirement for a similar first word in each of the first phrases.

```python
Parallelism1 = '''
chapter book=Esther
    c1:clause
        =: p1:phrase
        <: p2:phrase
        <: p3:phrase
        <: p4:phrase
    c2:clause
        =: pA:phrase
        <: pB:phrase
        <: pC:phrase
        <: pD:phrase

c1 <: c2
p1 .function=function. pA
p2 .function=function. pB
p3 .function=function. pC
p4 .function=function. pD
'''
Parallelism1  = BHS.search(Parallelism1)
BHS.table(Parallelism1, start=1, end=10, condensed=False, colorMap={1: 'purple', 2: 'orange',3: 'green',4: 'magenta',5: 'grey',6: 'yellow',7: 'orange',8: 'green',9: 'magenta',10: 'grey'})
```
